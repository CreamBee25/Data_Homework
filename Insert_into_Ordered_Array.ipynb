```python
class SortedArray:
    def __init__(self, initial_arr=None):
        if initial_arr is None:
            self._array = []
        else:
            self._array = initial_arr.copy()

   

    def _find_insert_index(self, element):
        """升序场景下的二分查找插入位置"""
        if not self._array:
            return 0  # 空数组直接插入到索引0

        left = 0
        right = len(self._array) - 1
        insert_index = len(self._array)  # 默认插入到末尾

        # 升序二分查找：找到第一个 >= element 的位置
        while left <= right:
            mid = (left + right) // 2
            if self._array[mid] >= element:
                insert_index = mid
                right = mid - 1
            else:
                left = mid + 1

        return insert_index

    def insert(self, element):
        """
        对外插入方法：升序插入元素（允许重复）
        :param element: 要插入的元素（数字/字符串等可比较类型）
        :return: 插入后的数组长度
        """
        insert_idx = self._find_insert_index(element)
        self._array.insert(insert_idx, element)
        return len(self._array)

   

    def __str__(self):
        """自定义打印格式：显示升序数组内容"""
        return f"SortedArray(升序, 允许重复): {self._array}"
```
